# DataFrame.plotとsubplotで柔軟にプロット

```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

## なんかサンプルデータつくる

```python
# Dailyのサンプルデータを生成
size = 25
df = pd.DataFrame(
{
    # カウントデータのサンプル
    'count1': np.random.randint(100, 150, size),
    'count2': np.random.randint(200, 250, size),
    'count3': np.random.randint(300, 350, size),
    'count4': np.random.randint(400, 450, size),
    
    # 比率データのサンプル
    'ratio1': np.random.rand(size).round(2),
    'ratio2': np.random.rand(size).round(2) / 2,
    'ratio3': np.random.rand(size).round(2),
    'ratio4': np.random.rand(size).round(2)  / 2
    },
    # indexは日付
    index = [ x.to_datetime().date() for x in pd.date_range(pd.datetime.today(), periods=size)]
)

# データ確認
df.head()
```

## プロットする

```python
# プロットするグラフの共通設定
cargs = {'kind':'line', 'use_index':True, 'ylim': (0,None), 'rot':45}

# グラフごとに個別に設定したいもの（プロットするカラムとか、タイトルとか）
args_list = [
    {'title': 'count A', 'column':['count1', 'count2']},
    {'title': 'count B', 'column':['count3', 'count4']},
    {'title': 'ratio A', 'column': ['ratio1', 'ratio2']},
    {'title': 'ratio B', 'column': ['ratio3', 'ratio4']},
]

# プロットの構成要素（レイアウトとか、axとか）取得
# 2×2のグラフでサイズは(20, 15), x軸はshare
fig, axes = plt.subplots(ncols=2, nrows=2, figsize=(20, 15), sharex=True)

# プロット
for ax, args in zip(axes.ravel(), args_list):
    df[args['column']].plot(ax=ax,  **cargs, title=args['title'])
```
